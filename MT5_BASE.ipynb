```python
from transformers import MT5ForConditionalGeneration, T5Tokenizer

# Load model and tokenizer
model_name = "google/mt5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = MT5ForConditionalGeneration.from_pretrained(model_name)

# French text to translate
french_text = "Le patient a de la fièvre et des maux de tête."

# Proper input formatting for MT5
input_text = f"translate French to English: {french_text}"

# Tokenize with proper parameters
inputs = tokenizer(
    input_text,
    return_tensors="pt",
    max_length=512,
    truncation=True,
    padding="max_length"
)

# Generate translation with adjusted parameters
outputs = model.generate(
    input_ids=inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_new_tokens=100,  # Increased from default
    num_beams=4,         # Better quality than greedy search
    early_stopping=True,
    no_repeat_ngram_size=2  # Avoids word repetition
)

# Proper decoding
translated_text = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True,
    clean_up_tokenization_spaces=True
)

print(f"French: {french_text}")
print(f"English: {translated_text}")
```

```python
from transformers import MT5ForConditionalGeneration, T5Tokenizer

# 1. Load model with correct settings
model_name = "google/mt5-small"
tokenizer = T5Tokenizer.from_pretrained(model_name, legacy=False)  # Important: legacy=False
model = MT5ForConditionalGeneration.from_pretrained(model_name)

# 2. Prepare text with PROPER prefix (critical for MT5)
french_text = "Le patient a de la fièvre et des maux de tête."
input_text = "translate French to English: " + french_text  # Must include this exact prefix

# 3. Tokenize with correct parameters
inputs = tokenizer(
    input_text,
    max_length=512,
    truncation=True,
    return_tensors="pt",
    padding="longest"  # Better than max_length for variable inputs
)

# 4. Generate with optimized settings
outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    num_beams=5,               # Better quality than greedy search
    early_stopping=True,
    no_repeat_ngram_size=2,    # Prevents word repetition
    temperature=0.7            # More natural output
)

# 5. Clean decoding
translated_text = tokenizer.decode(
    outputs[0],
    skip_special_tokens=True,
    clean_up_tokenization_spaces=True
)

print("French:", french_text)
print("English:", translated_text)
```

```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer
import torch
from nltk.translate.bleu_score import sentence_bleu
from nltk.tokenize import word_tokenize
import nltk

# Download NLTK data
nltk.download('punkt')

# Load mT5 model and tokenizer
model_name = "google/mt5-base"
tokenizer = MT5Tokenizer.from_pretrained(model_name)
model = MT5ForConditionalGeneration.from_pretrained(model_name)

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# French sentence to translate
french_sentence = "Le patient a ressenti une douleur thoracique sévère."

# Reference English translation
reference_translation = "The patient experienced severe chest pain."

# Format input for mT5 (prefixing with task name)
input_text = "translate French to English: " + french_sentence

# Tokenize input
inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True).to(device)

# Generate translation
outputs = model.generate(inputs, max_length=128, num_beams=4, early_stopping=True)
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Display translation
print("French Input:", french_sentence)
print("Translated Output:", translated_text)
print("Reference Translation:", reference_translation)

# BLEU score calculation
reference_tokens = [word_tokenize(reference_translation.lower())]
candidate_tokens = word_tokenize(translated_text.lower())
bleu_score = sentence_bleu(reference_tokens, candidate_tokens)

print("BLEU Score:", bleu_score)
```

```python
from transformers import MT5ForConditionalGeneration, MT5Tokenizer
import torch
from nltk.translate.bleu_score import sentence_bleu
import nltk

# Download NLTK BLEU dependencies (avoid punkt issue)
nltk.download('omw-1.4', quiet=True)

# Load MT5 model and tokenizer
model_name = "google/mt5-base"
tokenizer = MT5Tokenizer.from_pretrained(model_name)
model = MT5ForConditionalGeneration.from_pretrained(model_name)

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Russian input sentence and reference English translation
russian_sentence = "Пациент почувствовал сильную боль в груди."
reference_translation = "The patient felt severe chest pain."

# Input formatting (though mT5 isn’t fine-tuned for translation prompts)
input_text = "translate Russian to English: " + russian_sentence
inputs = tokenizer.encode(input_text, return_tensors="pt", truncation=True).to(device)

# Translate
outputs = model.generate(inputs, max_length=64, num_beams=4, early_stopping=True)
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Output
print("Russian Input:", russian_sentence)
print("Translated Output:", translated_text)
print("Reference Translation:", reference_translation)

# BLEU score
reference_tokens = [reference_translation.lower().split()]
candidate_tokens = translated_text.lower().split()
bleu_score = sentence_bleu(reference_tokens, candidate_tokens)

print("BLEU Score:", bleu_score)
```
